```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_mall=pd.read_csv('Mall_Customers.csv')
df_mall.head()
```

```python
df_mall.shape
```

# Exploratory Data Analysis

```python
df_mall=df_mall.drop(['CustomerID'], axis=1)
df_mall.head()
```

```python
age=df_mall['Age'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(age.index, age.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
Gender=df_mall['Genre'].value_counts()
plt.figure(figsize=(12, 8))
sns.barplot(Gender.index, Gender.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df_mall.head()
```

```python
income=df_mall['Annual Income (k$)'].value_counts()
plt.figure(figsize=(20,4))
sns.barplot(income.index, income.values)
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.show()
```

```python
spending=df_mall['Spending Score (1-100)'].value_counts()
plt.figure(figsize=(20, 6))
sns.barplot(spending.index, spending.values)
plt.xlabel('Spending Score')
plt.ylabel('Count')
plt.show()
```

```python
df_mall.isnull().sum()
```

```python
df_mall.columns=['Gender', 'Age', 'Annual Income', 'Spending Score']
df_mall.head()
```

```python
sns.pairplot(df_mall, hue='Gender')
plt.show()
```

# Univariate Analysis

```python
fig, ax=plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
print(df_mall['Gender'].value_counts())
sns.countplot(x=df_mall['Gender'], ax=ax[0, 0])
sns.distplot(df_mall['Age'], bins=50, ax=ax[0, 1])
sns.distplot(df_mall['Annual Income'], bins=50, ax=ax[1, 0])
sns.distplot(df_mall['Spending Score'], bins=50, ax=ax[1, 1])
plt.show()
```
